```python
import nltk
import string
string.punctuation
import re
   
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
#pd.set_option('max_columns', 120)
pd.set_option('display.max_colwidth',200)
dataset=pd.read_json (r'C:\Users\Profveca\Desktop\projet\test.json')
#df=pd.DataFrame(dataset)
#print(df)
dataset.head()
```

```python
stopWords = set(stopwords.words('english'))

def cleaning_dataset(txt):
    txt="".join([i for i in txt if i not in string.punctuation])
    tokens=re.split('\W+', txt)
    ques=[word for word in tokens if word not in stopWords]
    lemmatizer=WordNetLemmatizer()
    lemmatized_output =[lemmatizer.lemmatize(word.lower()) for word in ques if word ]
   # tagged = nltk.pos_tag(lemmatized_output)

    return lemmatized_output
dataset['questions_clean']=dataset['questions'].apply(lambda x: {"question": cleaning_dataset(x["question"]), 
                                                                 "post_tag": x["post_tag"]})


dataset.head()
```

```python
### on a creer trois listes une pour question ,une pour pos_tag(intents) et l'autre pour couple intent et POS_TAG(NN,adv,aj,etc)

X = list(map(lambda x: " ".join(x["question"]), dataset['questions_clean']))
Y =list(map(lambda x: x["post_tag"], dataset['questions_clean']))
Xpos = list(map(lambda x: nltk.pos_tag(x["question"]), dataset['questions_clean']))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#CountVectorizer to count the number of words (term frequency),limit your vocabulary size, apply stop words###
                                                                                
cv=CountVectorizer()
x=cv.fit_transform(X)
print(x.shape)
df=pd.DataFrame(x.toarray(),columns=cv.get_feature_names())
df.head()

#print(cv.get_feature_names())
```

```python
# Split dataset into training set and test set

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3) # 70% training and 30% test

 
#per = 0.3
#nbtest = int(len(X)*per)
#X_test = X[:nbtest]
#X_train = X[nbtest:]
#Y_test = Y[:nbtest]
#Y_train = Y[nbtest:]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
#from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
#from sklearn.linear_model import SGDClassifier
#from sklearn.naive_bayes import MultinomialNB

import numpy as np



model_pipeline = Pipeline(steps=[
        ('vect', CountVectorizer()),
        ('tfidf', TfidfTransformer()),
        #('dtc', DecisionTreeClassifier()), accuracy =28.7
       #('mul', MultinomialNB()),accuracy =22
       ('rfc', RandomForestClassifier()),
      # ('sgdc', SGDClassifier())  accuracy =36
                            ])




rfcc = model_pipeline.fit(X_train,Y_train)

#Predict the response for test dataset
y_pred = model_pipeline.predict(X_test)

print("Accuracy:\n",accuracy_score(Y_test, y_pred)*100)
print("Report :\n ",classification_report(Y_test, y_pred)) 
```
