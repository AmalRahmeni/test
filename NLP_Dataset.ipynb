```python
import nltk
import string
string.punctuation
import re
   
    
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
pd.set_option('max_columns', 120)
pd.set_option('display.max_colwidth',150)
dataset=pd.read_json (r'C:\Users\Profveca\Desktop\projet\test.json')
#df=pd.DataFrame(dataset)
#print(df)
dataset.head()
```

```python
##remove punctuation 

def remove_punctuation(txt):
    txt_nopunct="".join([i for i in txt if i not in string.punctuation])
    return txt_nopunct

dataset['questions_clean']=dataset['questions'].apply(lambda x: {"question": remove_punctuation(x["question"]), 
                                                                 "post_tag": x["post_tag"]})
dataset.head()
```

```python
##tokenize

def tokenize(txt):
    tokens=re.split('\W+', txt)
    return tokens
dataset['questions_clean_tokenized']=dataset['questions_clean'].apply(lambda x: {"question": tokenize(x["question"]), 
                                                                                 "post_tag": x["post_tag"]})
dataset.head()




```

```python
##remove stop Words

stopWords = set(stopwords.words('english'))
#print(stopWords)
def remove_stopWords(txt_tokenized):
    questions_clean=([word for word in txt_tokenized if word not in stopWords])
    return questions_clean
dataset['questions_no_sw']=dataset['questions_clean_tokenized'].apply(lambda x:{"question": remove_stopWords(x["question"]), 
                                                                                 "post_tag": x["post_tag"]})
dataset.head()
```

```python
##lemmatization

def lemmatization(token_txt):
    #text=[wn.lemmatize(word) for word in token_txt]
    lemmatizer=WordNetLemmatizer()

    lemmatized_output = ' '.join ([lemmatizer.lemmatize(word.lower()) for word in token_txt])
    return lemmatized_output

dataset['questions_lemmatized']=dataset['questions_no_sw'].apply(lambda x:{"question": lemmatization(x["question"]), 
                                                                                 "post_tag": x["post_tag"]})

dataset.head()
```
