```python
import requests
from bs4 import BeautifulSoup
import json
import time
p = 0
questions_data = {
    "questions":[]
}
while len(questions_data['questions'])<10000:
    p += 1

    res = requests.get("https://stackoverflow.com/questions?tab=newest&pagesize=50&page="+str(p))

    soup = BeautifulSoup(res.text, "html.parser")

    questions = soup.select(".question-summary")

    for que in questions:

        q = que.select_one('.question-hyperlink').getText()
        tag = que.select_one('.post-tag').getText()
       
        quest={
            "question": q,
            "post_tag":tag,
        }
        
        questions_data['questions'].append(quest)
            
    

json_data = json.dumps(questions_data , indent=2)

print(json_data)

open("test.json","w",encoding="utf-8").write(json_data)
     
    
```
